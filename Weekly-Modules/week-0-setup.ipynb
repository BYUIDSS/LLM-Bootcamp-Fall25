# 2. Clone the git repository, open in VS Code

The curriculum is hosted on Github, we need to move it to our local machines

### What is Github?

A git repository is a codebase hosted on the service Github. It allows many people to work on the same code by having a local version of the codebase on your computer, and a public version that notes who changed what. In our case we are only pulling, or taking the code from Github and moving it to our local machines so we can run it.

Cloning a repo is the process of copying the code from Github to your machine.

## Step by Step

There are several ways to do this, but this is for someone that isn't familiar with cloning a repo.

### VS Code Method (generally easiest)

1. Open VS Code

2. Press Ctrl + Shift + P (or Cmd + Shift + P on Mac) to open the Command Palette

3. Type "Git: Clone" and press Enter

4. Paste this URL and hit enter: https://github.com/BYUIDSS/LLM-Bootcamp-Fall25.git

    - You can access the Github project page online by visiting this website https://github.com/BYUIDSS/LLM-Bootcamp-Fall25

5. Select a folder on your computer where you want the project to live.

6. A notification will appear in the bottom right; click Open to start working on the code immediately.

### Alternative: Download as ZIP

If you don't want to deal with Git or GitHub accounts at all and just want the files once:

1. On the GitHub project page, click the green <> Code button.

2. Select Download ZIP.

3. Extract the folder and open it in VS Code via File > Open Folder...

Note: Downloading the ZIP is faster for a one-time look, but you won't be able to easily update the code later or send your changes back to GitHub.

# 3. Additional Steps

You may or may not encounter these issues

Issues with `pip` or Python

1. Make sure you have the Python extension (left hand side of VS Code or `ctrl + shift + x`)

2. Try running `pip --version` and `python --version`

3. If one of them looks off, uninstall and reinstall Python (not through Windows store)

4. If they look fine, close all instances of VS Code and reopen

Issues with Jupyter Notebooks

1. Make sure you have the Jupyter extension

For any other issues, ask AI or a neighbor
